# Analyzing Education Attainment Across Generations
- There are some pretty big differences between age groups
- Can analyze in depth more for sure

```python
import matplotlib.pyplot as plt
import pandas as pd

def edu_dot_chart(file):
    # headers on row 5
    df_title = pd.read_excel(file, header=None)
    # dynamically printing title
    print("\033[1m" + "Title: " + df_title.iloc[3, 0] + "\033[0m")

    df = pd.read_excel(file, header=5)

    # ":," means select all rows, 0 is for the column
    # this just gets rid of the tabs in the first column because they look weird when doing df.head()
    # turning regex on means it'll find the \t anywhere in the string
    df.iloc[:, 0] = df.iloc[:, 0].replace("\t", "", regex=True) 

    # 0 is a description, 1 and 3 are an aggregate of age groups
    df = df.drop([0, 1, 3])

    df = df.iloc[0:12].copy()
    df.set_index(df.columns[0], inplace=True)

    df.replace('Z', 0, inplace=True)
    # exclude the first two columns and plot the rest
    df = df.iloc[:, 2:].transpose()
    
    # split the dataframe for plotting
    young_df = df.iloc[:, 0:6]
    old_df = df.iloc[:, 6:]

    plt.figure(figsize=(10, 5))
    for c in young_df.columns:
        plt.scatter(young_df.index, young_df[c], label=c)
    
    # show mean
    means_young = young_df.mean(axis=1)
    for i, mean in enumerate(means_young):
        plt.plot([means_young.index[i]], [mean], 'r_', markersize=12, label='Mean' if i == 0 else "")

    plt.xlabel("Educational Attainment")
    plt.ylabel("Total (Thousands)")
    plt.title("Educational Attainment by Age Group (First Half)")
    plt.legend(title="Age Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 5))
    for c in old_df.columns:
        plt.scatter(old_df.index, old_df[c], label=c)
    
    # show mean
    means_old = old_df.mean(axis=1)
    for i, mean in enumerate(means_old):
        plt.plot([means_old.index[i]], [mean], 'r_', markersize=12, label='Mean' if i == 0 else "")

    plt.xlabel("Educational Attainment")
    plt.ylabel("Total (Thousands)")
    plt.title("Educational Attainment by Age Group (Second Half)")
    plt.legend(title="Age Group")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

for i in range(1, 10):
    edu_dot_chart(f"Table1/table-1-{i}.xlsx")
```
